```python
from Bio.PDB import PDBList

pdbl = PDBList()

# Download PDB files for amylase, trypsin, and lipase
pdbl.retrieve_pdb_file('1AMY', pdir='.')
pdbl.retrieve_pdb_file('1TRN', pdir='.')
pdbl.retrieve_pdb_file('1LFP', pdir='.')
```

```python
import nglview as nv
from Bio.PDB import MMCIFParser

# Parse the downloaded PDB files
parser = MMCIFParser()
amylase_structure = parser.get_structure('amylase', '1amy.cif')
trypsin_structure = parser.get_structure('trypsin', '1trn.cif')
lipase_structure = parser.get_structure('lipase', '1lfp.cif')

# Create NGL viewer objects
view_amylase = nv.show_biopython(amylase_structure)
view_trypsin = nv.show_biopython(trypsin_structure)
view_lipase = nv.show_biopython(lipase_structure)

# Display the 3D structures
print('Amylase')
display(view_amylase)

print('Trypsin')
display(view_trypsin)

print('Lipase')
display(view_lipase)
```
